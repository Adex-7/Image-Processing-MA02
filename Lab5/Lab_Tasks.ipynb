```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Task#1: Convolve an image with a 7x7 box filter, and output the original and smoothed image side by side
 
img = cv2.imread('parrot.jpg')
 
kernel = np.ones((7,7),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Smoothed')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

#Task2: Using an image of your choice, apply a 5x5 gaussian filter to the image, then a 21x21 gaussian filter and show the three images side by side
 
img = cv2.imread('parrot.jpg')
 
blur_5 = cv2.GaussianBlur(img,(5,5),0)
blur_21 = cv2.GaussianBlur(img,(21,21),0)

# Display the images
plt.figure(figsize=(10, 5))  # Create a figure with a specified size
plt.subplot(1, 3, 1)  # Subplot for the original image
plt.imshow(img)
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)  # Subplot for the 5 x 5 blur image
plt.imshow(blur_5)
plt.title('5 x 5')
plt.axis('off')

plt.subplot(1, 3, 3)  # Subplot for the  21 x 21 blur image
plt.imshow(blur_21)
plt.title('21 x 21')
plt.axis('off')
plt.show()
```

```python
# Task#3: Using an image of your choice, apply a 3x3 Laplacian filter to sharpen the image. 
image = cv2.imread('cat.1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_float = image.astype(np.float32) / 255.0

# Use the cv2.Laplacian() function
# We use cv2.CV_32F as the depth since our image is float32, and ksize=3 for a 3x3 filter.
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

# Sharpen the image
sharpened_image = np.clip(image_float - laplacian, 0, 1)

# Plot Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_float)
plt.title('1. Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
# Adding 0.5 to shift the zero-point to gray for easier viewing
plt.imshow(np.clip(laplacian + 0.5, 0, 1)) 
plt.title('2. Laplacian Filter')
plt.axis('off')

# Plot Sharpened Image
plt.subplot(1, 3, 3)
plt.imshow(sharpened_image)
plt.title('3. Sharpened Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
