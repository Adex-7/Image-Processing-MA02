```python
#Add or remove libraries as needed
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

### Task#1: Load/read cat.jpeg as a grayscale image [1 mark]

```python
# loading the image as gray
img = cv2.imread('cat.jpeg', cv2.IMREAD_GRAYSCALE)
#showing the image
plt.imshow(img, cmap=cm.gray)
plt.title('orignal image')
plt.axis('off')
plt.show()
```

### Task#2: Apply thresholding (thresh=100) to the image and display the results [1.5 mark]
#### Note: use cmap = 'gray' to properly display the thresholded image

```python
# applaying the thresholding to the image and make it = 100
img_thres = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
#showing the image side by side with the original image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap=cm.gray)
plt.title('orignal image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_thres[1], cmap=cm.gray)
plt.title('thresholed image')
plt.axis('off')
plt.show()
```


### Task#3: Load/read the camera man, blur it using a 9x9 gaussian filter (sigma=5), and display the results.   [2.5 marks]

```python
# load the image
img = cv2.imread('cameraman.tif')
# apply blur to the image with sigma 5
img_blur = cv2.GaussianBlur(img, (9, 9), 5)
# showing the image side by side with the original image

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('orignal image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_blur, cmap='gray')
plt.title('blurr image')
plt.axis('off')
plt.show()
```
