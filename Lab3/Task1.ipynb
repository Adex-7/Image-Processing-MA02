```python
import cv2
img = cv2.imread('cat.1.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale image', gray_img)
cv2.waitKey()
```

```python
yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
#To separate out the three channels, write the following lines:
cv2.imshow('Y channel', yuv_img[:, :, 0])
cv2.imshow('U channel', yuv_img[:, :, 1])
cv2.imshow('V channel', yuv_img[:, :, 2])
cv2.waitKey()
```

```python
#Task1 
import matplotlib.pyplot as plt
#1. Increase the size of the image

img_enlarged = cv2.resize(img, None,fx=1.5, fy=1.5,interpolation=cv2.INTER_CUBIC)

#2: Rotate the image by 120 degrees.

import numpy as np
img = cv2.imread('cat.1.jpg', 0)
rows, cols = img.shape
M = np.float32([[1,  0, 0], [0, -1, rows], [0,  0, 1]])

img_rotation = cv2.warpAffine(img,cv2.getRotationMatrix2D((cols/2, rows/2),120, 0.6),(cols, rows))


#3. Perform Sheer operations over the image.

rows, cols = img.shape
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))


# Display the images
plt.figure(figsize=(10, 5))  # Create a figure with a specified size
plt.subplot(1, 4, 1)  # Subplot for the original image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("original_image")

plt.subplot(1, 4, 2)  # Subplot for the original image
plt.imshow(cv2.cvtColor(img_enlarged, cv2.COLOR_BGR2RGB))
plt.title("img_enlarged")

plt.subplot(1, 4, 3)  # Subplot for the original image
plt.imshow(cv2.cvtColor(img_rotation, cv2.COLOR_BGR2RGB))
plt.title("img_rotation")
plt.axis('off')
plt.subplot(1, 4, 4)  # Subplot for the original image
plt.imshow(cv2.cvtColor(sheared_img, cv2.COLOR_BGR2RGB))
plt.title("sheared_img")
plt.axis('off')
plt.show()
```
